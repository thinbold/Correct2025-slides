## Wiretap

```sage
def apply(x, step):
    if not step: return x
    s, tep = step[0], step[1:]
    if s == 'p': x = x^2
    if s == 's': x = 1 - (1 - x)^2
    return apply(x, tep)
```

```sage
step = 'pppssss'
apply(0.1, step), 1 - apply(0.9, step)
```

```sage
step = 'pspspspsps'
apply(0.1, step) < 0.01, 1 - apply(0.9, step) < 0.001
```

## Quantum

```sage
def ten(x, y):
    return vector([xi * yi for xi in x for yi in y])
```

```sage
ten([1, 4], [1, 2])
```

```sage
p = vector([1, 1])
m = vector([1, -1])
pp = ten(p, p)
pm = ten(p, m)
mp = ten(m, p)
mm = ten(m, m)
ppp = ten(pp, p)
ppm = ten(pp, m)
pmp = ten(pm, p)
pmm = ten(pm, m)
mpp = ten(mp, p)
mpm = ten(mp, m)
mmp = ten(mm, p)
mmm = ten(mm, m)    
```

```sage
def mat(x, y):
    return matrix(x).T * matrix(y)
```

```sage
mat(p, m) + mat(m, p)
```

```sage
mat(pp, pp) + mat(pm, pm) + mat(mp, mm) + mat(mm, mp)
```

```sage
mat(ppp, ppp) + mat(ppm, ppm) + mat(pmp, pmp) + mat(pmm, pmm) + \
    mat(mpp, mpp) + mat(mpm, mpm) + mat(mmp, mmm) + mat(mmm, mmp)
```
