```python
def helpdata(q):
    F = GF(q, repr='log')
    g = F.gen()
    mp = g.minpoly()
    print(f'$\FF_{{{q}}}$ is generated by', end=' ')
    print(f'{''.join(str(j) for j in reversed(list(mp)))} (i.e., ${mp}$).')
    print(r'Its Zech table (i.e., $1 + \alpha^j = \alpha^{z(j)}$) is \\')

    for j in range(q):
        print(F(1 + g^j), end='')
        if j != q - 1: print(', ', end='')
        else: print('.')
        if (j + 1) % 10 == 0: print(r'\\')
    
    print()

helpdata(4)
```

```python
helpdata(4)
helpdata(8)
helpdata(16)
helpdata(32)
helpdata(64)
helpdata(128)
helpdata(256)
```
